```python
src = 'v:/clips'
dst = 'v:/images'
```

```python
import os
folders = os.listdir(src)
folders.sort()
```

```python
import ffmpeg
from tqdm.notebook import tqdm
import os

if not os.path.exists(dst):
    os.makedirs(dst)
for fd in tqdm(folders):
    if not os.path.exists(os.path.join(dst,fd)):
        os.makedirs(os.path.join(dst, fd))
    files = os.listdir(os.path.join(src, fd).replace('\\','/'))
    files.sort()
    startnm = 0
    for f in tqdm(files):
        print(f'{os.path.join(src, fd, f)}')
        ffmpeg.input(os.path.join(src, fd, f), ss = 0, r = 6)\
        .output(os.path.join(dst, fd, f'{fd}_%6d.jpg'), start_number=startnm, r = 1/10.5)\
        .overwrite_output()\
        .run(quiet=True)
        # startnm = int(os.listdir('./frames/tf2')[-1].split('.')[0][-6:])+1
        startnm = len(os.listdir(os.path.join(dst, fd)))
```

```python
def convert_mp4_to_h264_and_play_vdo(file_startwith='TF2202308301454'):
    import subprocess
    import os
    from IPython.display import Video

    clips = os.listdir('/home/fsd/FlareDetect/yolov7/SaveResult_Clips_tf2')
    clips.sort(reverse=True)
    for clip in clips:
        if file_startwith in clip:
            vdo = clip
    print(vdo)
    clip = vdo[:-4]
    input_path = f'/home/fsd/FlareDetect/yolov7/SaveResult_Clips_tf2/{clip}.mp4'
    output_path = f'/home/fsd/FlareDetect/yolov7/saveclip2h264/{clip}_h264.mp4'

    ffmpeg_command = [
        'ffmpeg',
        '-v', 'error',  # Suppress verbose output
        '-y',  # Overwrite output files without asking
        '-i', input_path,
        output_path
    ]

    # Run the FFmpeg command
    subprocess.run(ffmpeg_command)

    print(f"{clip} Conversion completed.")
    return output_path
convert_mp4_to_h264_and_play_vdo('TF2202308301658')

from IPython.display import Video
clip = convert_mp4_to_h264_and_play_vdo()
Video(clip)
```
