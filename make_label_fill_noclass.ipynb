```python
import os
import shutil
from tqdm.notebook import tqdm
```

```python
dir = 'V:/data_annotation/tf1_000000-006000'
```

```python
file_list = os.listdir(dir+'/images')
included_extensions = ['jpg','jpeg', 'bmp', 'png', 'gif']
files = [fn for fn in file_list
        if any(fn.endswith(ext) for ext in included_extensions)]
files.sort()
files = files[:]
files
```

```python
for f in tqdm(files):
    shutil.copyfile(f'{dir}/images/seaview2_002007.txt', f'{dir}/images/{f[:-3]}txt')
```

```python
for f in tqdm(files):
    if not os.path.exists(f'{dir}/labels/{f[:-3]}txt'):
        shutil.copyfile(f'{dir}/labels/seaview_noclass.txt', f'{dir}/labels/{f[:-3]}txt')
        print(f'{f[:-3]}txt')
```
